```python
# Iterate over each image and export it
def export_image(image_id, index):
    image = ee.Image('NOAA/GOES/18/MCMIPF/' + image_id).toInt16()
    clipped_image = image.clip(roi)

    # Export the clipped image to Google Drive
    task = ee.batch.Export.image.toDrive(
        image=clipped_image,
        description=f'clipped_GOES_image_{index}',
        folder='GEE_exports_permian_basin',  # Replace with your preferred Google Drive folder
        fileNamePrefix=f'{image_id}',
        region=roi_geometry.getInfo()['coordinates'],  # Define the export region
        scale=2000,  # Adjust scale according to your needs
        maxPixels=1e13,  # Set to a high number to allow larger exports
        crs='EPSG:4326'  # Set to the desired coordinate reference system
    )
    task.start()
    print(f'Exporting image {index} to Google Drive.')
```

```python
# Define the collection
collection = ee.ImageCollection('NOAA/GOES/18/MCMIPF')

# Get the current date and time using Python's datetime module
now = datetime.utcnow()
end_date = ee.Date(now.strftime('%Y-%m-%dT%H:%M:%S'))
start_date = end_date.advance(-10, 'year')  # 24 hours ago

# Define your region of interest (ROI) - replace `table` with your actual region
roi = ee.FeatureCollection('projects/ee-seismosmsr/assets/permian_basin')

# Use the geometry of the first feature in the FeatureCollection as the ROI
roi_geometry = roi.geometry()

# Filter the collection by date and region
filtered_collection = collection.filterDate(start_date, end_date).filterBounds(roi)

# Get a list of image IDs in the filtered collection
image_ids = filtered_collection.aggregate_array('system:index').getInfo()


# Specify the folder where TIFF files are saved
download_folder = '/content/drive/MyDrive/GEE_exports_permian_basin'

# List all existing TIFF files in the folder
existing_files = [f.split('.')[0] for f in os.listdir(download_folder) if f.endswith('.tif')]

# Filter image IDs to exclude already downloaded files
filtered_image_ids = [image_id for image_id in image_ids if image_id not in existing_files]

# Use `filtered_image_ids` instead of `image_ids` for further processing
print('Image IDs:', filtered_image_ids)
print(len(filtered_image_ids))
```

```python
from ee.batch import Task
from collections import Counter
import time

# Function to get the current number of active tasks
def count_active_tasks():
    tasks = Task.list()
    task_status_list = [task.status()['state'] for task in tasks[0:250]]
    status_counts = Counter(task_status_list)
    active_tasks = status_counts.get('READY', 0) + status_counts.get('RUNNING', 0)
    return active_tasks

# Maximum tasks you want to maintain
MAX_TASKS = 100
WAIT_INTERVAL = 3 * 60  # 3 minutes

# Iterate over the filtered images and export them
index = 0
while index < len(filtered_image_ids):
    # Get the number of active tasks
    active_tasks = count_active_tasks()

    # Determine how many new tasks to add
    tasks_to_add = MAX_TASKS - active_tasks
    if tasks_to_add > 0:
        for _ in range(tasks_to_add):
            if index < len(filtered_image_ids):
                export_image(filtered_image_ids[index], index)
                index += 1
            else:
                break  # Exit if there are no more images to export

    # Wait for 3 minutes before checking again
    print(f"Waiting for 3 minutes, current active tasks: {active_tasks}")
    time.sleep(WAIT_INTERVAL)
```
